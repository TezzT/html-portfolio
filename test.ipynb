```python
import os
import requests

# List of font URLs
font_urls = [
    "https://static.jjwxc.net/tmp/fonts/jjwxcfont_6gyzt.eot?h=my.jjwxc.net",
    "https://static.jjwxc.net/tmp/fonts/jjwxcfont_6gyzt.woff2?h=my.jjwxc.net",
    "https://static.jjwxc.net/tmp/fonts/jjwxcfont_6gyzt.woff?h=my.jjwxc.net",
    "https://static.jjwxc.net/tmp/fonts/jjwxcfont_6gyzt.ttf?h=my.jjwxc.net",
    "https://static.jjwxc.net/tmp/fonts/jjwxcfont_6gyzt.svg?h=my.jjwxc.net"
]

# Directory to save the fonts
font_directory = "fonts"
os.makedirs(font_directory, exist_ok=True)

# Download each font and save to the local directory
for url in font_urls:
    try:
        # Send a GET request to fetch the font file
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful

        # Extract the file name from the URL
        font_name = url.split('/')[-1].split('?')[0]

        # Save the content to a local file
        with open(os.path.join(font_directory, font_name), 'wb') as file:
            file.write(response.content)

        print(f"Successfully downloaded: {font_name}")

    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")
```

```python
def extract_gsub_mappings(font_path):
    font = TTFont(font_path)
    gsub_table = font["GSUB"].table
    mappings = {}

    for lookup in gsub_table.LookupList.Lookup:
        for subtable in lookup.SubTable:
            if hasattr(subtable, "mapping"):  # Direct substitutions
                mappings.update(subtable.mapping)
    
    return mappings

# Extract glyph substitutions
gsub_mappings = extract_gsub_mappings("New folder\\fonts\\jjwxcfont_6h25q.ttf")

# Print mappings
for k, v in gsub_mappings.items():
    print(f"U+{k:04X} -> U+{v:04X}")
```

```python
from fontTools.ttLib import TTFont

font = TTFont("New folder/fonts/jjwxcfont_6h25q.ttf")

# Extract character map (PUA to glyph names)
cmap = font["cmap"].getcmap(3, 1).cmap  # Unicode cmap
pua_mapping = {code: name for code, name in cmap.items() if 0xE000 <= code <= 0xF8FF}

print("PUA Character Mappings:")
for code, glyph in pua_mapping.items():
    print(f"U+{code:04X} -> {glyph}")
```

```python
from bs4 import BeautifulSoup

# Load the SVG font file
with open("New folder\\fonts\\jjwxcfont_6h25q.svg", "r", encoding="utf-8") as file:
    svg_data = file.read()

# Parse XML
soup = BeautifulSoup(svg_data, "xml")

# Extract glyph mappings
glyphs = soup.find_all("glyph")

for glyph in glyphs:
    unicode_char = glyph.get("unicode")
    glyph_name = glyph.get("glyph-name")
    print(f"{unicode_char} -> {glyph_name}")
```

```python
import uharfbuzz as hb

# Open TTF as binary
with open("jjwxcfont_6h25q.ttf", "rb") as f:
    font_data = f.read()

# Load font into HarfBuzz
hb_face = hb.Face(font_data)  
hb_font = hb.Font(hb_face)

# Example PUA text
text = "‌‌‌"  # Replace with your actual PUA characters

# HarfBuzz shaping
buffer = hb.Buffer()
buffer.add_str(text)
buffer.guess_segment_properties()
hb.shape(hb_font, buffer)

# Extract glyph info
glyph_info = buffer.glyph_infos
glyph_positions = buffer.glyph_positions

# Print shaped output
for i, g in enumerate(glyph_info):
    print(f"Glyph {i}: id={g.codepoint}, cluster={g.cluster}, x_advance={glyph_positions[i].x_advance}, y_advance={glyph_positions[i].y_advance}")
```

```python
from fontTools.ttLib import TTFont

# Load the font file (change the path to your actual file)
font_path = "New folder/fonts/jjwxcfont_6h25q.woff2"
font = TTFont(font_path)

# Extract character mappings
cmap = font.getBestCmap()

# Save mappings to a file
with open("font_mappings.txt", "w", encoding="utf-8") as f:
    for codepoint, glyphname in cmap.items():
        char = chr(codepoint)
        f.write(f"U+{codepoint:04X} -> {glyphname} ({char})\n")

print("✅ Glyph mappings extracted! Check font_mappings.txt")
```
